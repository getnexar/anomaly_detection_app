```python
import requests
import json
from typing import List, Dict, Optional

class SimpleVectorSearch:
    def __init__(self, api_url: str = "https://data-platform-api.prod-or.k.nexar.mobi"):
        self.api_url = f"{api_url}/api/v1/video/search"
        self.headers = {'Content-Type': 'application/json'}
    
    def get_embedding(self, video_id: str) -> Optional[List[float]]:
        """Get embedding vector for a video by searching and finding it"""
        # Try multiple search approaches to find the video with embedding
        search_queries = ['video', 'car', 'driving', 'road', 'vehicle', 'incident', '']
        
        for query in search_queries:
            try:
                # Search with different queries and large result set
                response = requests.post(self.api_url, json={
                    'free_text_search': query,
                    'extended_results': True,
                    'k_value': 100  # Get more results to increase chances
                }, headers=self.headers, timeout=30)
                
                if response.status_code != 200:
                    continue
                    
                results = response.json().get('results', [])
                
                # Find our video in results
                for result in results:
                    if result.get('video_id') == video_id:
                        embedding = result.get('embedding')
                        if embedding:
                            print(f"✅ Found embedding using query: '{query}'")
                            return embedding
                        
            except Exception as e:
                continue
        
        print(f"❌ Could not find video {video_id} with embedding in any search")
        return None
    
    def search_similar(self, embedding: List[float], limit: int = 10) -> List[Dict]:
        """Search for similar videos using embedding"""
        response = requests.post(self.api_url, json={
            'embedding_search': embedding,
            'k_value': limit
        }, headers=self.headers)
        
        if response.status_code != 200:
            return []
            
        return response.json().get('results', [])
    
    def text_search(self, query: str, limit: int = 10) -> List[Dict]:
        """Regular text search"""
        response = requests.post(self.api_url, json={
            'free_text_search': query,
            'k_value': limit
        }, headers=self.headers)
        
        if response.status_code != 200:
            return []
            
        return response.json().get('results', [])


def demo():
    # Initialize client
    client = SimpleVectorSearch()
    
    print("🔍 Vector Search Demo\n")
    
    # 1. Get videos with embeddings directly
    print("1. Getting videos with embeddings...")
    response = requests.post(f"{client.api_url}", json={
        'free_text_search': 'car',
        'extended_results': True,
        'k_value': 5
    }, headers=client.headers)
    
    if response.status_code != 200:
        print("❌ Failed to get videos")
        return
    
    videos = response.json().get('results', [])
    
    # Find a video that has an embedding
    sample_video = None
    for video in videos:
        if video.get('embedding'):
            sample_video = video
            break
    
    if not sample_video:
        print("❌ No videos with embeddings found")
        return
    
    video_id = sample_video['video_id']
    description = sample_video.get('description', 'No description')
    embedding = sample_video['embedding']
    
    print(f"✅ Sample video: {video_id}")
    print(f"   Description: {description[:80]}...")
    print(f"   Embedding: {len(embedding)} dimensions\n")
    
    # 2. Find similar videos using the embedding
    print("2. Finding similar videos...")
    similar_videos = client.search_similar(embedding, limit=5)
    
    if not similar_videos:
        print("❌ No similar videos found")
        return
    
    print(f"✅ Found {len(similar_videos)} similar videos:\n")
    
    for i, video in enumerate(similar_videos, 1):
        vid_id = video.get('video_id', 'Unknown')
        desc = video.get('description', 'No description')
        similarity = video.get('similarity', 0)
        distance = video.get('distance', 0)
        
        print(f"{i}. Video: {vid_id}")
        print(f"   Description: {desc[:60]}...")
        print(f"   Similarity: {similarity:.4f} | Distance: {distance:.4f}")
        print()


def find_similar_to_video(video_id: str, num_results: int = 5) -> List[Dict]:
    """Simple function to find videos similar to a given video ID"""
    client = SimpleVectorSearch()
    
    # Get embedding
    embedding = client.get_embedding(video_id)
    if not embedding:
        print(f"Could not get embedding for {video_id}")
        return []
    
    # Search similar
    return client.search_similar(embedding, num_results)


def search_and_find_similar(text_query: str, num_similar: int = 3) -> None:
    """Search by text, then find videos similar to the first result"""
    client = SimpleVectorSearch()
    
    print(f"🔍 Searching for: '{text_query}'")
    
    # Search with extended_results to get embedding directly
    response = requests.post(f"{client.api_url}", json={
        'free_text_search': text_query,
        'extended_results': True,
        'k_value': 10
    }, headers=client.headers)
    
    if response.status_code != 200:
        print("❌ Search failed")
        return
    
    videos = response.json().get('results', [])
    
    # Find first video with embedding
    base_video = None
    for video in videos:
        if video.get('embedding'):
            base_video = video
            break
    
    if not base_video:
        print("❌ No videos with embeddings found")
        return
    
    print(f"📹 Base video: {base_video['video_id']}")
    print(f"   {base_video.get('description', 'No description')[:80]}...\n")
    
    # Find similar using the embedding we already have
    embedding = base_video['embedding']
    similar = client.search_similar(embedding, num_similar)
    
    if similar:
        print(f"🎯 Similar videos:")
        for i, video in enumerate(similar, 1):
            print(f"{i}. {video['video_id']} (similarity: {video.get('similarity', 0):.3f})")
            print(f"   {video.get('description', 'No description')[:60]}...")
            print()
    else:
        print("❌ No similar videos found")
```
